```python
import matplotlib.pyplot as plt
```

```python
product1 = [10,20,30,40,50]
product2 = [11,220,330,450,540]
month = [1,2,3,4,5]
```

```python
plt.plot(month,product1,marker=".",linestyle="--")
plt.plot(month,product2,marker="o",linestyle="-.")
plt.xlabel("month")
plt.ylabel("sales")
plt.savefig("export1.png",transparent=True)
```
