```python
import matplotlib.pyplot as plt
```

```python
course = ["py","jv","c"]
score = [80,75,50]
c= ["green", "orange","blue"]
```

```python
fig, ax = plt.subplots()

ax.barh(course,score,color=c,edgecolor="black" , linewidth = 3)
ax.set_xlabel("Course")
ax.set_ylabel("score")
ax.set_title("Score Midterm")
plt.show()
```
