```python
import matplotlib.pyplot as plt
```

```python
product1 = [10,20,30,40,50]
product2 = [11,220,330,450,540]
month = [1,2,3,4,5]
```

```python
plt.plot(month,product1,marker=".")
plt.plot(month,product2,marker="o")

data={"size":20,"color":"red","backgroundcolor" :"yellow"}

plt.xlabel("month",data)
plt.ylabel("sales",data)
config={"size":20,"color":"blue"}
plt.title("Top sale 2020",config,loc="left")
plt.savefig("export1.png",transparent=True)
plt.text(1.9,230,"point",size="xx-large",color ="green")
```
