```python
import matplotlib.pyplot as plt
```

```python
data = [15,30,45,60,40]
course = ["PHY","PY","JV","C","Node"]

```

```python
fig, ax = plt.subplots()
color = ["purple","red","cyan","yellow","green"]
exp = [0,0.2,0,0.2,0]
ax.pie(data, labels=course,autopct="%.2f%%",colors=color,shadow=True,explode=exp,startangle=45)
plt.show()
```
