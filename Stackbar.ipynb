```python
import matplotlib.pyplot as plt
```

```python
room = ["A","B","C"]
boys = [10,15,5]
girls = [20,7,15]
```

```python
plt.bar(room,boys,label = "Boys",color = "blue")
plt.bar(room,girls,bottom = boys,label="girls",color="orange")
plt.legend()
```

```python
plt.stackplot(room,boys,girls,labels=["boys","girls"],colors=["green","pink"])
plt.legend()
```
